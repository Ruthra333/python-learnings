wap for numbers lesser than 50 check if they are divisible by 2,for numbers between 50 to 100 check if they are divisible by 3,for numbers greater than 100 check if they are divisible by 7

```python
num=int(input('enter a number'))
print('num =',num)
if num<50:
  if num%2==0:
    print('the num is less than 50 and divisible by 2')

if 50<=num<=100:
  if num%3==0:
    print('the num is between 50 to 100 and divisible by 3')

if num>100:
  if num%7==0:
    print('the num is greater than 100 and divisible by 7')

else:
  print('condition not met')
```

```python
num = int(input('Enter a number: '))
print('num =', num)

if num < 50:
    if num % 2 == 0:
        print('The num is less than 50 and divisible by 2')
    else:
        print('The num is less than 50 and not divisible by 2')
elif 50 <= num <= 100:
    if num % 3 == 0:
        print('The num is between 50 to 100 and divisible by 3')
    else:
        print('The num is between 50 to 100 and not divisible by 3')
elif num > 100:
    if num % 7 == 0:
        print('The num is greater than 100 and divisible by 7')
    else:
        print('The num is greater than 100 and not divisible by 7')
else:
    print('Condition not met')
```
