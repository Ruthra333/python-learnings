Here we have two lists (year and GDP). Create a line plot as proper labels and title to the figure.
Year = [1950, 1960, 1970, 1980, 1990]
GDP = [50.8, 65.9, 75.8, 92.6, 125.8]

```python
Year = [1950, 1960, 1970, 1980, 1990]
GDP = [50.8, 65.9, 75.8, 92.6, 125.8]

sns.lineplot(x=Year,y=GDP,color="red")
plt.text(x=1965,y=72,rotation=22,s="gdp_line")
plt.xlabel("Year")
plt.ylabel("GDP")
plt.title("GDP growth rate")
```

Here we have two lists (year and GDP) add marker, color, size and line width.
Year = [1950, 1960, 1970, 1980, 1990]
GDP = [50.8, 65.9, 75.8, 92.6, 125.8]

```python
sns.lineplot(x=Year,y=GDP,color="red",lw=2,marker='o',ms=30)                             # ms --> size of marker
plt.xlabel("Year")
plt.ylabel("GDP")
plt.title("GDP growth rate")
```

Use mtcars dataset and create a scatter plot between mpg (mile per gallon) and wt (weight) feature with proper lables and title. Change the marker style to hexagon, add face color as blue, edge color as Navy and size 40.

Use mtcars dataset and create a scatter plot between mpg (mile per gallon) and wt (weight). Scatter plot should be between am 0 and 1.

```python
sns.scatterplot(x=u["mpg"],y=u['wt'],hue=u["am"],marker="H",facecolor="blue",edgecolor="navy",s=40)
plt.xlabel("MPG")
plt.ylabel("weight")
plt.title("correlation of mpg vs weight")


#---or---

am0=u[u['am']==0]                                                                                         # we create two separate data frames based on the am value and plot two scatter plot in one

am1=u[u['am']==1]

plt.scatter(x=am0["mpg"],y=am0['wt'],marker="H",facecolor="blue",edgecolor="navy",s=40,label="am0")

plt.scatter(x=am1["mpg"],y=am1['wt'],marker="D",facecolor="red",edgecolor="navy",s=40,label="am1")

plt.legend()
plt.xlabel("MPG")
plt.ylabel("weight")
plt.title("correlation of mpg vs weight")
```
