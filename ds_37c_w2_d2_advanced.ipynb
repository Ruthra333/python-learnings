DATETIME

```python
import datetime
```

```python
datetime.datetime.now()                                                                  # current date and time
```

```python
datetime.datetime.now().time()                                                           # only time

datetime.datetime.now().date().strftime('%d-%m-%y')                                      # only date
```

```python
x=datetime.datetime.now().date()                                                         # formatting

x.strftime('%D-%M-%Y')                                                                   # 01-01-2024

x.strftime('%d-%m-%y')                                                                   # 03-01-24

x.strftime("%d-%B-%y")                                                                   # 03-Jan-24       # ** %h ** can be used for the same    # B --> january

x.strftime("%d-%h-%y")
```

```python
x+datetime.timedelta(days=10)                                                      # to +,-, of date

x-datetime.timedelta(days=10)                                                      # minus can be added before int of days  -10

x+datetime.timedelta(weeks=52,seconds=999,minutes=999,hours=9)

datetime.timedelta
```

```python
x.weekday()                                                                         # to get day of the week      monday=0--sunday=6

x.strftime("%a")                                                                    # both same

import calendar

calendar.day_name[4]    # or    # calendar.day_name[x.weekday()]
```

```python
# to get diff in date if one is give ex:2015-04-09

x=datetime.datetime.now()

y=datetime.datetime(2015,4,9,1,15,22,2345)

r=x-y

r.days
```

REGULAR EXPRESSIONS

```python
import re
```

```python
s="helloworld123"

x=re.match('[a-z]+',s)                                                             # will return value only if found at the beginning of the string
                                                                                      # []-->for setting a range ,  + for more than one values

x
```

```python
s="123helloworld123"

x=re.match('[a-z]+',s)                                                             # it is not returning any value as the beginning value doesnt match

x

if x:                                                                              # if x has value it will print if statement and vice versa   similar to true or false
  print("found")

else:
  print("not found")
```

```python
re.search(r'\b[a-z]{5}\b','the jessa and kelly')                                   # finds the first match and stops

re.search(r'\b[a-zA-Z]{5}\b','the Jessa and kelly')                                # the diff is  it takes both small and caps of words with 5 letters

re.search(r'\b[a-zA-Z]{5}\b\Z','the Jessa and Kelly')                              # \b is for single word ,  \Z for matching at the end of sentences.

re.findall(r'\b[a-zA-Z]{5}\b','the Jessa and Kelly')                               # will provide all with five letters
```

```python
re.findall(r'\b[^a-zA-Z]{5}\b','the Jessa and Kelly')                              # no output bcoz we put ^(means - not) before it

re.findall(r'\b[^a-zA-Z]{5}\b','the Jessa 12345 and Kelly')

re.findall(r'\d','the Jessa and  12345 Kelly')                                     # OTHER \W,\w,\s,\Z,\a,\b,\d
```

```python
re.split(' ',"hello world 123")                                                   # split based on space ' '

re.split('\.',"hello.world.123")                                                  # *** as .  already has meaning in regex we use \ to negate and say it is normal . ***
```

```python
re.sub('h','b',"helloworld123")                                                   # to substitute one in place of other
```
