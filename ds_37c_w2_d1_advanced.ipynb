MERGING AND JOINING OF DATAFRAMES

```python
df1=pd.DataFrame({'a':[2,4,23,4,6,2],'b':['d','f','ws','e','w','y']})

df1
```

```python
df3 =pd.DataFrame({'b':['d','f',230,40,60,200,300],'d':['dasd','fg','wsgh','eaf','aew','hyj','fg']})

df3
```

```python
pd.merge(df1,df3)                                                                    #by default it is **inner join**
```

```python
pd.merge(df1,df3,how="left")                                                        # it does left join & similarly for right join

pd.merge(df1,df3,how="right")
```

```python
pd.merge(df1,df3,how="outer")                                                     # merges  all the values
```

```python
df1.join(df3)                                                             # as two column(b) has same name it do not know how to join as join works on index and (also checks for column names)

```

to achive join where 2column has same name

```python
df1.set_index('b').join(df3.set_index('b')).reset_index()                            # we taking **b**  which is repeated and setting it as index in both so the column gets removed and changed to index an then we join and reset it to get back normal 1b not 2 b's
# left join bvy default


df1.set_index('b').join(df3.set_index('b'),how='inner').reset_index()


df1.set_index('b').join(df3.set_index('b'),how='outer').reset_index()

df1.set_index('b').join(df3.set_index('b'),how='right').reset_index()
```
