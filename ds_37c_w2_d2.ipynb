From the string "There was a great king. His name was Arthur. He ruled many years", make 2 separate lists. First, containing all words with length less than 4 converted to all lowercase. Second, containing all words with length >= 4, with all converted to upper case. The fullstops should not be part of the result.

```python
s="there was a great king.his name was arthur.he ruled many years"

for items in s:
  if len(s)<4:
    print(items.lower())
  if len(s)>=4:
    print(items.upper())
```
