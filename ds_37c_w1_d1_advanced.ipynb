recursive function revision


```python
def fact(n):
  if n==1:
    return 1
  else:
    return n*fact(n-1)

fact(5)
```

**decorator function**

```python
def my_deco(duck):                        # this is how we create a decorator 7 now it is ready to be used for any functiopn in place of duck
  def wrapper():
    print("start")                         # my own example
    duck()
    print("end")
  return wrapper
```

```python
@my_deco
def muck():
  print("bonjour")

muck()
```

```python
import datetime
```

```python
def my_deco(func):
  def wrapper(x):
    print(datetime.datetime.now())
    func(x)
    print(datetime.datetime.now())
  return wrapper
```

```python
@my_deco
def my_func(x):
  print("hellow",x)
my_func('ruth')
```

fibanacii using recursive

```python
def fib(n):
  if n<=1:
    return n
  else:
    return(fib(n-1)+fib(n-2))

print(fib(11))
```
