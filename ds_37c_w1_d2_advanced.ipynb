**Concatinating and staking of arrays**

```python
y1=np.array([[4,85,7],[5,25,69]])

y2=np.array([[5,1,4],[1,56,6]])
```

```python
y1
```

```python
y2
```

```python
np.concatenate([y1,y2])                             # row wise -- no of DIMENSION IS SAME AS ORIGINAL ARRAYS

np.concatenate([y1,y2],axis=1)                      # column wise -- no of DIMENSION IS SAME AS ORIGINAL ARRAYS
```

```python
np.stack([y1,y2])                                  # DIMENSION INCREASES
```

```python
np.hstack([y1,y2])                                #same as concantenate with axis 1               #dimension doesnt get changed
```

```python
np.vstack([y1,y2])                                # stacks based on the position            #dimension doesnt get changed    axis=0
```
