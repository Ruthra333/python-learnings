WAP to check if numbers in a list are even or odd.

```python
my_list = [7,4,67,3,897,22,46]

for i in my_list:
  if i%2==0:
    print (i,"is even")
  else:
    print (i,'is odd')
```

WAP to change the value for key "Name3" to "Gaurav". And replace all values that are of length 3 with "AAA" in the below dictionary. {'Name1':'AB','Name2':'BC','Name3':'CD','Name4':'def','Name5':'cdf'}

```python
d = {'Name1':'AB','Name2':'BC','Name3':'CD','Name4':'def','Name5':'cdf'}
d['Name3']="gaurav"

for keys,values in d.items():
  if len(d[keys])==3:
    d[keys]='AAA'
print(d)
```

```python
d = {'Name1':'AB','Name2':'BC','Name3':'CD','Name4':'def','Name5':'cdf'}

for keys in d.keys():
  if keys=='Name3':
    d[keys]='Gaurav'
  if len(d[keys])==3:
    d[keys] = 'AAA'
print(d)
```
