```python
#positional arguments -----> we have to provide the value based on the given position example as follows

def sum_num(num1,num2):
  print("first number is",num1)
  print("second number is",num2)
  print("sum =",num1+num2)

sum_num(8,2)#--------->the no inside the argument automatically takes the position of num1 & 2


#in case we want to know which one is for which one and also in case we do not know the position of numbers we want to insert then you can do as follows

#default arguments----->explanation as follows
sum_num(num2=98,num1=49)


#in case if you forgot to enter value inside the created function as supposed to ....then it throws error to avoid error we can code as follows

def sum_num(num1=0,num2=0): #------>by providing values in case we not enter values it automatically takes the intially provided values
  print("first number is",num1)
  print("second number is",num2)
  print("sum =",num1+num2)

sum_num()
```

```python
def sum_of_numbers(*arg):#variable-length arguments---->arg is used for getting n number of input
  sum=0
  for i in arg:
    sum=sum+i
  print("sum=",sum)

sum_of_numbers(10,2,0,1,6350,150,51)

sum_of_numbers(1)
```

```python
def var_len_Key_arg(**karg):
  for key,val in karg.items():
    print(key,"has value",val)


var_len_Key_arg(name='ruth',age=23,sex='male')
```

WAF that takes any number of integers and tells wheteher the sum of integer is odd or even

```python
def func(*arg):
  sum=0
  for i in arg:
    sum=sum+i
  print("the sum is",sum)
  if sum%2==0:
    print("sum is even")
  else:
    print("suum is odd")

func(2,265,3,5,56)
```

to add if list is given along with normal values like 2,3,51,[5,96,8,2],984,554,5------->to add all values ignoring the bracket

```python
def func(*arg):
  sum=0
  for i in arg:
    if isinstance(i,list):
      for j in i:
        sum=sum+j
    else:
      sum=sum+i
  print("the sum is",sum)

func( 2,3,[5,8,2],5)

```

saving the output in the return function and why is it  used? as follows

```python

def sum_num(num1,num2):
  print("first number is",num1)
  print("second number is",num2)
  print("sum =",num1+num2)      #here the print function simply print the output but you are not storing the output value for any further processes
  return num1+num2               # so we are using return statement to store the output
sum_num(2,5)

# all inbuilt function has return statement in it for eg:r=len() it gives result based on return statement


# can also return more than one

def sum_num(num1,num2):
  print("first number is",num1)
  print("second number is",num2)
  print("sum =",num1+num2)
  return num1+num2,num1-num2
sum_num(2,5)

# can also assign variable to returned value

sum,diff=(num1+num2),(num1-num2)
```
