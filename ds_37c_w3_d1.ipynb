```python
l=[1,5,6,25,4,2,54,1,54,116,545,56]

#shallow copy

l1=l   #pointing to the same memory location ------> shallow copy

l1[0]=999

l1

l #here both l1 & l are altered by changing any one of them


#deep copy

l2=l.copy() #pointing to different memeory location------>deep copy

l2[2]=5555

l2

l #here altering l2has no effect on l

# --------------or--------------
l2=l[:]                                # it is also deepl copy has no effect on each other

l2[5]=9898

l2
l
```

```python
d=dict()
type(d)
d

d['a']=1
d
d['a']=10

d.update({'b':20,'c':30})        #to insert multiple values in dictionary
d



for i in d.keys():
  print(i,'--->',d[i])




li=[1,2,34,5,67,896]

dic=dict.fromkeys(li,77)              #  function to assign values to all keys at a time...........but cannot give multiple values.
dic

f=('ruthr',"tuhrA")

fof=dict.fromkeys(f,"smooth")

fof





dd={1:20,2:200,3:3000,4:40000}        # using deep copy in an dictionary.......has no effect on original

d=dd.copy()

d[4]=40

d
```
