parent and child class

```python
class person():                                                                     # this is an parent class
  def __init__(self,name,age):
    self.name=name
    self.age=age
  def display(self):                                                               # this second method is just for viewing purpose (optional)
    print("parent class has been called")
    print("name :",self.name,"Age :",self.age)
  def __str__():
    print("instead of default message it can give you the output of your choice")
```

```python
c1=person("dhoni","41")

c1.display()
```

```python
class employee(person):                                                              # this is a child class inheriting from parent class (subclass/derivedclass)  as  it gets is info from parent class named person also this works as same a parent (i.e) it also requiers to attributes
  def disp(self):
    print("this is a child class")
```

```python
c2=employee("thala","14")                                                          # creating obj for child class   # __init__ from parent class is called automatically
c2.disp()
```

```python
c2.display()                                                                      # using same object we can call function of parent class here it is (display)
```

```python
class employee2(person):
  def display(self):                                                            # here we are creating another child but the twist is we will name the function identical to parent function  (display)
    print(" over rided parent class has been called")                           # in this case it will **overwrite** it with its(child) function/method
```

```python
c3=employee2("thalapathy",45)

c3.display()
```

Task 7
Create a class called Animal with the following specifications:

Attributes:
species (protected attribute denoted by a single underscore prefix)
breed (private attribute denoted by double underscore prefix)

Methods:

Create methods to describe the sound and activity of Animal.

Note: Give the input in the code itself

Expected output:
Name:  Leo
Species:  Lion
The animal makes a sound.
Name:  Buddy
Species:  Dog
Breed:  Labrador
The dog barks.
The dog wags its tail.

```python
#code goes from here
class Animal():
    def __init__(self, name, species):
        self.name = name
        self._species = species
        print("Name: ", self.name)
        print("Species: ", self._species)

    def set_sound(self, sound):
        self.sound = sound

    def get_activity(self):
        print("Generic animal activity.")

class Lion(Animal):
    def get_activity(self):
        print("The animal makes a sound.")

class Dog(Animal):
    def set_breed(self, breed):
        self.breed = breed
        print("Breed: ", self.breed)

    def get_activity(self):
        print("The dog barks.")
        print("The dog wags its tail.")

animal = Lion("Leo", "Lion")
animal.get_activity()

dog = Dog("Buddy", "Dog")
dog.set_breed("Labrador")
dog.get_activity()
```
